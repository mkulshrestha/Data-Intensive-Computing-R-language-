```r
data=read.csv("FluViewPhase2Data-3/WHO_NREVSS_Clinical_Labs.csv", TRUE, ",", skip=1)
par(bg="white")
pasted_value =(sprintf("%d%02d", data$YEAR,data$WEEK))

barplot(data$TOTAL.A, main="Influenza positive tests reported by CDC US Clinincal Lab \nNational Summary, season 2017-2019",ylim=c(0,max(data$TOTAL.A)),ylab="No. of +ve specimens",xlab="WEEKS",cex.axis=0.8,names.arg = pasted_value, col="yellow",las=2,cex.names=0.5)

par(new=T)

barplot(data$TOTAL.B,axes=FALSE, ylim=c(0, max(data$TOTAL.B)), col="green")

par(new=T)

plot(data$PERCENT.POSITIVE, axes=FALSE, type="l",ylab="",xlab="")
axis(4, las=1,at=seq(0, max(data$PERCENT.POSITIVE),by=5))
mtext("Percent positive", side=4, line=1)
lines(data$PERCENT.A, type="l",col="yellow",lty="dashed")
lines(data$PERCENT.B, type="l",col="green",lty="dashed")

legend("topright",legend=c("percent positive","percent A", "percent B","A","B"),
       col=c("black","yellow", "green","yellow","green"),cex=0.8,lty=c(1,3,3,0,0),pch=c(NA,NA,NA,15,15))
```

```r
data2=read.csv("FluViewPhase2Data-3/WHO_NREVSS_Public_Health_Labs.csv",header=TRUE,",",skip=1)
par(bg="white")

pasted_value =(sprintf("%d%02d", data2$YEAR,data2$WEEK))

data2=data.frame(data2$BYam,data2$BVic,data2$B,data2$H3N2v,data2$A..H3,data2$A..2009.H1N1, 
                 data2$A..Subtyping.not.Performed)

data2=t(data2)
# data2
barplot(as.matrix(data2), main="INfluenza positive tests reported by CDC US labs \n National Summary, season 2017-2019",col=c("green","light green"," dark green","purple", "red","orange","yellow"),names.arg = pasted_value, las=2, cex.names=0.5)

legend("topright",legend=c("A subtyping not performed","A 2009 H1N1","A H3N2","H3N2V","B non lineage","B Victorian",
                           "B Yamagata"),
       col=c("yellow","orange","red","purple","dark green","light green","green"),
       cex=0.8,
       lty=c(0,0,0,0,0,0,0),pch=c(15,15,15,15,15,15,15))
```

```r
data3=read.csv("FluViewPhase2Data-4/WHO_NREVSS_Clinical_Labs.csv",TRUE,",",skip=1)
par(bg="white")
pasted_value =(sprintf("%d%02d", data3$YEAR,data3$WEEK))
barplot(data3$TOTAL.A, xlab = "weeks", ylab="percent positive",ylim=c(0,max(data3$TOTAL.A)),names.arg = pasted_value, col="yellow",axes=FALSE)
axis(2, at=seq(0,max(data3$TOTAL.SPECIMENS),by=100))

par(new=TRUE)

bg=barplot(data3$TOTAL.B,axes=FALSE, xlab="weeks", ylab="percent positive", ylim=c(0, max(data3$TOTAL.B)),names.arg = pasted_value, col="green")

par(new=TRUE)

plot(data3$PERCENT.POSITIVE, axes=FALSE, type="l",ylab="",xlab="")
axis(4, las=1,at=seq(0, max(data3$PERCENT.POSITIVE),by=5))
lines(data3$PERCENT.A, type="l",col="yellow",lty="dashed")
lines(data3$PERCENT.B, type="l",col="green",lty="dashed")

legend("topright",legend=c("percent positive","percent A", "percent B","A","B"),
       col=c("black","yellow", "green","yellow","green"),cex=0.8,lty=c(1,3,3,0,0),pch=c(NA,NA,NA,15,15))
```
